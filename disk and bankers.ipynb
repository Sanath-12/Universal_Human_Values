```python
import matplotlib.pyplot as plt
import seaborn as sns
cylinders = [10,22,20,2,40,6,38]
start = 20
# considering disk drive has 41 cylinders( numbered 0-40)
cylinder_begin = 0
cylinder_end = 40
seek_time= 6
```

```python
def fcfs(cylinders,start,seek_time):
  print(cylinders)
  cylinders.insert(0,start)
  level= [32,28,24,20,16,12,8,4]
  plt.plot(cylinders,level)
  sns.scatterplot(x=cylinders, y=level, color='black')
  for i in range(len(cylinders)):
    plt.annotate(f'({cylinders[i]})', xy=(cylinders[i], level[i]))
  plt.xlabel('cylinders')
  plt.ylabel('')
  plt.gca().set_yticklabels([])
  plt.title("FCFS")
  total_movement = 0
  for i in range(0,len(cylinders)-1):
    total_movement += abs(cylinders[i]-cylinders[i+1])
  print(f"Total Movement = {total_movement}")
  print(f"Total Seek Time = {total_movement*seek_time}")

fcfs(cylinders,start,seek_time)
```

```python
def c_scan(cylinders,start,cylinder_begin,cylinder_end):
  level= [32,28,24,20,16,12,8,4]
  if cylinder_end not in cylinders:
    cylinders.append(cylinder_end)
    level.insert(0,level[0]+4)
  if cylinder_begin not in cylinders:
    cylinders.append(cylinder_begin)
    level.insert(0,level[0]+4)
  cylinders.sort()
  cylinders = cylinders[cylinders.index(start):] + cylinders[0:cylinders.index(start)] 
  cylinders.insert(0,start)
  print(cylinders)
  plt.plot(cylinders,level)
  sns.scatterplot(x=cylinders, y=level, color='black')
  for i in range(len(cylinders)):
    plt.annotate(f'({cylinders[i]})', xy=(cylinders[i], level[i]))
  plt.xlabel('cylinders')
  plt.ylabel('')
  plt.gca().set_yticklabels([])
  plt.title("C-SCAN")
  total_movement = 0
  for i in range(0,len(cylinders)-1):
    total_movement += abs(cylinders[i]-cylinders[i+1])
  print(f"Total Movement = {total_movement}")
  print(f"Total Seek Time = {total_movement*seek_time}")

c_scan(cylinders,start,cylinder_begin,cylinder_end)
```
